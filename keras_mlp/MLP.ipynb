## Libraries

```python
from sklearn.preprocessing import MinMaxScaler

import keras
from keras.models import Sequential
from keras.utils import np_utils
from keras.layers.core import Dense, Activation, Dropout
from keras.optimizers import RMSprop

import pandas as pd
import numpy as np

np.set_printoptions(precision=4)
np.set_printoptions(suppress=True)
```

## Prepare data

```python
def prepare_data(traindata, testdata):
    numerics = ['int16', 'int32', 'int64', 'float16',
            'float32', 'float64']
    traindata = traindata.select_dtypes(include=numerics)
    testdata = testdata.select_dtypes(include=numerics)
    traindata.fillna(0, inplace=True)
    testdata.fillna(0, inplace=True)
    train_input_matrix = traindata.values[:,1:37]
    train_output_matrix = traindata.values[:,37]
    train_output_matrix = train_output_matrix.reshape(-1,1)
    test_input_matrix  = testdata.values[:,1:]
    normalized_train_input_matrix = scaler_input_features.fit_transform(train_input_matrix)
    normalized_test_input_matrix = scaler_input_features.transform(test_input_matrix)
    normalized_train_output_matrix = scaler_saleprice.fit_transform(train_output_matrix)
    return normalized_train_input_matrix, normalized_test_input_matrix, normalized_train_output_matrix
```

## MLP parameters

```python
OUTPUT_FILENAME = '../data/keras/result_al.csv'

NR_NEURONS_HIDDEN1 = 20
NR_NEURONS_HIDDEN2 = 10
NR_NEURONS_OUTPUT  = 1

NR_TRAIN_STEPS = 100000
LEARN_RATE = 0.001

#min_max_scaler.inverse_transform(scaled_data)
```

## Setup MLP

```python
def setup():
    model = Sequential()
    model.add(Dense(160, activation="relu"))
    model.add(Dense(80, activation="relu"))
    model.add(Dense(1, activation="linear"))
    return model
```

## Training the mlp

```python
def train(model, train_matrix, train_output): 
    model.compile(loss='mean_squared_error', optimizer='sgd')  
    model.fit(train_matrix, train_output, validation_split=0.10, epochs=1000)
    model.summary()
    return model
```

## Predict the house prices

```python
def predict(model, test_matrix):
    print("Generating test predictions...")
    preds = model.predict(test_matrix)
    return preds, scaler_saleprice.inverse_transform(preds)
```

## Write data to CSV file

```python
def write_to_csv(test_data, pred_in_dollar):
    test_house_ids = test_data.values[:,0] # get the IDs from the original Pandas DataFrame
    pred_in_dollar = pred_in_dollar.reshape(-1)
    print("test_house_ids has shape", test_house_ids.shape)
    print("preds_test_houses_dollar has shape", pred_in_dollar.shape)
    prediction_dataframe = pd.DataFrame({'Id'       :test_house_ids,
                                        'SalePrice':pred_in_dollar}
                                      )

    # convert column "Id" to int64 dtype
    prediction_dataframe = prediction_dataframe.astype({"Id": int})

    prediction_dataframe.to_csv(OUTPUT_FILENAME, sep=',', index=False)
```

## Main

```python
traindata = pd.read_csv('../data/train.csv')
testdata = pd.read_csv('../data/test.csv')
scaler_input_features = MinMaxScaler(feature_range=(0, 1))
scaler_saleprice = MinMaxScaler(feature_range=(0, 1))

train_matrix, test_matrix, train_output = prepare_data(traindata, testdata)
mlp = setup()
trained_mlp = train(mlp, train_matrix, train_output)

predictions, pred_in_dollar = predict(trained_mlp, test_matrix)
print(pred_in_dollar)
write_to_csv(testdata, pred_in_dollar)
```
